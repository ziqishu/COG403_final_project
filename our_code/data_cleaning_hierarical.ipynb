```python

def sample_from_large_file(file_path, sample_size, chunksize=10000):
    # Step 1: Determine the number of rows
    total_rows = sum(1 for _ in open(file_path, 'r', encoding='utf-8')) - 1  # Subtract 1 for the header
    print(f"Total rows in the file: {total_rows}")  # Debugging
    
    # Calculate the fraction to sample from each chunk
    fraction = sample_size / total_rows
    print(f"Sampling fraction: {fraction}")  # Debugging
    
    # Placeholder for sampled rows
    sampled_df = pd.DataFrame()
    rows_sampled = 0
    
    # Step 2: Read the file in chunks and sample
    for chunk in pd.read_csv(file_path, chunksize=chunksize, encoding='utf-8'):
        # Calculate number of rows to sample from this chunk, ensuring at least 1 row is sampled
        rows_to_sample = max(1, int(fraction * len(chunk)))
        sampled_chunk = chunk.sample(n=rows_to_sample)
        sampled_df = pd.concat([sampled_df, sampled_chunk], ignore_index=True)
        rows_sampled += len(sampled_chunk)
        
        # Stop if we've collected enough samples
        if rows_sampled >= sample_size:
            break
    
    return sampled_df.head(sample_size) 
```

```python
sampled_df = sample_from_large_file(file_path, sample_size)
```
