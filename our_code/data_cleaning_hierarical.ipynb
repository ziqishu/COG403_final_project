```python
import pandas as pd

# Load the shared response survey data with specific columns (ExtendedSessionID and Review_political)
df_shared_response_survey = pd.read_csv('SharedResponsesSurvey.csv', usecols=['ExtendedSessionID', 'Review_political', 'Review_gender'])

# Convert df_shared_response_survey into a dictionary for efficient lookup
review_political_dict = pd.Series(df_shared_response_survey.Review_political.values,index=df_shared_response_survey.ExtendedSessionID).to_dict()
review_gender_dict = pd.Series(df_shared_response_survey.Review_gender.values,index=df_shared_response_survey.ExtendedSessionID).to_dict()

# Initialize an empty list to store the data
data = []
header_modified = False
n_matched_rows = 0  # Counter for matched rows

with open("C:/Users/Ziqi/OneDrive - University of Toronto/cog403-ZiqiShu/project_data/Datasets/Moral Machine Data/SharedResponses.csv") as f:
    for i, row in enumerate(f):
        if not header_modified:
            # Process header
            header = row.strip().split(',') + ['Review_political'] + ['Review_gender']
            data.append(header)  # Add header as the first row in the list
            header_modified = True
            continue
        
        row_data = row.strip().split(',')
        extended_session_id = row_data[1]  # Adjust the index as necessary
        
        # Use the dictionary to find Review_political for the ExtendedSessionID
        if extended_session_id in review_political_dict:
            # Append the Review_political value to the row data and add it to the list
            row_data.append(review_political_dict[extended_session_id])
            row_data.append(review_gender_dict[extended_session_id])
            data.append(row_data)
            n_matched_rows += 1
            
            if n_matched_rows % 10 == 0:  # Print progress every 10 matched rows
                print(f"Matched {n_matched_rows} rows")
            
            # Stop after including 5000 matched rows
            if n_matched_rows >= 5000:
                break

# Convert the list of data into a DataFrame, excluding the first row which is the header
df = pd.DataFrame(data[1:], columns=data[0])

print("Data processing complete.")

# Optional: Save the processed DataFrame to a new CSV file
# df.to_csv("FilteredSharedResponses.csv", index=False)
```

```python
df['Review_political_cat'] = pd.cut(df['Review_political'], bins=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=[1, 2, 3, 4, 5])
```

```python
# randomly select 3000 rows from df
df_sample1 = df.sample(n=3000, random_state=1)
df_sample2 = df.sample(n=3000, random_state=2)

print(df_sample1['Review_gender'].value_counts())
print(df_sample2['Review_gender'].value_counts())

print(df_sample1['Review_political_cat'].value_counts())
print(df_sample2['Review_political_cat'].value_counts())

df_sample1.to_csv("data_ml/SharedResponses_sample1.csv", index=False)
df_sample2.to_csv("data_ml/SharedResponses_sample2.csv", index=False)
```

```python
# randomly select 1232 rows of males (in Review_gender Column) and 1232 rows of female in two different dataframe
df_male = df[df['Review_gender'] == 'male'].sample(n=1232, random_state=1)
df_female = df[df['Review_gender'] == 'female'].sample(n=1232, random_state=1)

# for each df, count the number of each unique Review_political
print(df_male['Review_political_cat'].value_counts())
print(df_female['Review_political_cat'].value_counts())

df_male.to_csv("data_ml/SharedResponses_male.csv", index=False)
df_female.to_csv("data_ml/SharedResponses_female.csv", index=False)

# combine the two dataframes df_male and df_female
df_combined = pd.concat([df_male, df_female], ignore_index=True)
df_combined.to_csv("data_ml/SharedResponses_combined_two_genders.csv", index=False)
```
